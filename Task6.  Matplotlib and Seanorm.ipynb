```python
# Scatter Plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.scatter(x, y)
plt.title('Simple Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

Interpretation : we see a positive correlation between x and y as the values of x increase as y increases.

```python
# Bar Plot : Bar charts are used to compare different categories or groups.
fruits = ['Apple', 'Banana', 'Cherry', 'Mango']
counts = [10, 20, 30, 40]
plt.bar(fruits, counts)
plt.title('Fruit Count')
plt.xlabel('Fruit Type')
plt.ylabel('Count')
plt.show()
```

Interpretation :bar chart shows the count of different fruits.observe that Mango has the highest count, 
    followed by Cherry, Banana, and Apple.

```python
# Histogram : Histograms are used to show the distribution of data
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

Interpretation : histogram shows the distribution of 1000 randomly generated data points.The data is normally distributed, as the histogram has a bell-shaped curve.

```python
# pie chart :Pie charts are used to show the proportion of categories within a whole.
labels = ['Apple', 'Banana', 'Cherry', 'Mango']
sizes = [10, 20, 30, 40]
plt.pie(sizes, labels=labels)
plt.title('Fruit Distribution')
plt.show()
```

Interpretation :pie chart shows the distribution or praportation of different fruits in terms of percentage.mango makes up the largest portion, followed by Cherry, Banana, and Apple.

```python
# Box Plot :  show the distribution of data, highlighting the median, quartiles, and outliers.
data = [np.random.normal(0, 1, 100), np.random.normal(1, 1.5, 100), np.random.normal(2, 1, 100)]
plt.boxplot(data)
plt.title('Box Plot of Data')
plt.xlabel('Group')
plt.ylabel('Value')

# Show plot
plt.show()
```

```python
#Stacked Bar Chart : Stacked bar charts are used to show the total values and the subgroup of different categories.
categories = ['A', 'B', 'C', 'D']
data1 = [10, 15, 20, 25]
data2 = [5, 10, 15, 20]
plt.bar(categories, data1, label='Data1', color='blue')
plt.bar(categories, data2, bottom=data1, label='Data2', color='orange')
plt.title('Stacked Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.legend()
plt.show()
```

Interpretation : This chart displays two datasets Data1 and Data2 stacked on top of each other for each category. We can see that the total value for each category is the sum of Data1 and Data2, and seprated by the colour.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill",y="tip",data=tips)

sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time", style="sex", size="size")

plt.title("Scatter Plot")
plt.show()


#Hue:group data points by a categorical variable(color) ie Colors the points based on the time variable either "Lunch" or "Dinner"
#style:group by a categorical variable and change the marker style
```

1. plot will show that as the total bill increases, the ti also increases, indicating a positive relationship between these two variables.
2. Time of DAy for Lunch or dinner different colors will show there is a noticeable difference in tips between lunch and idnner.
3. The size of the markers increases that indicate the larger parrties tend to leave higher tips.

```python
# Bar Plt(seaborn.barplot) :

sns.barplot(x="day", y="total_bill", data=tips)
sns.barplot(x="day", y="total_bill", data=tips, hue="sex")

plt.title('Bar Plot')
plt.show()
```

```python
# Histogram

sns.histplot(tips['total_bill'], kde=True)

sns.histplot(tips['total_bill'], kde=True, bins=20, color="blue", multiple="stack")

plt.title('Histogram with KDE')
plt.show()

# kde: Include Kernel Density Estimate (smooth curve).
# bins: Number of bins in the histogram.
# color: Specify color for the bars.
```

Interpretation :  histogram shows the distribution of total_bill with a smooth KDE curve.

```python
# box plot
sns.boxplot(x="day", y="total_bill", data=tips)
sns.boxplot(x="day", y="total_bill", data=tips,hue="sex",  palette="Set2")

plt.title('Box Plot')
plt.show()

# palette="Set2": The palette parameter controls the color scheme of the plot.
# hue="sex": This parameter splits the data within each category.
```

Interpretation : plot will show two box plots for each dayff and comparison between the distributions of total bills for each sex.

```python
# Violin Plot(seaborn.violinplot)

sns.violinplot(x="day", y="total_bill", data=tips)

sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)

plt.title('Violin Plot')
plt.show()
# hue: Group by categorical variable and color the violins.
#split: If True, splits the violins for different hue categories.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset("tips")
corr = tips.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot: total_bill vs tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

Interpretation : This scatter plot shows the relationship between total_bill and tip. The positive correlation indicates that as the total bill increases, the tip amount also tends to increase.


```python
# Box plot: tip by sex
sns.boxplot(x='sex', y='tip', data=tips)
plt.title('Box Plot: Tip by Sex')
plt.show()
```

Interpretation : This box plot shows the distribution of tip for each sex category. It reveals that there is little difference in the median tip between males and females. However, there are some outliers, especially for male.

```python
# Violin plot: tip by smoker
sns.violinplot(x='smoker', y='tip', data=tips)
plt.title('Violin Plot: Tip by Smoker')
plt.show()
```

INterprtation : The violin plot shows the distribution of tips based on smoker status. Smokers tend to leave slightly higher tips than non-smokers, and the spread of tips is also broader for smokers.


```python
# Pair plot: Pairwise relationships between variables
sns.pairplot(tips, hue='sex')
plt.title('Pair Plot: Pairwise Relationships')
plt.show()
```

```python
# Bar plot: Average tip by day
sns.barplot(x='day', y='tip', data=tips)
plt.title('Bar Plot: Average Tip by Day')
plt.show()
```

Interpretation : The bar plot shows the average tip by day. We observe that the tips are generally higher on weekends (Saturday and Sunday) compared to weekdays (Thursday and Friday), possibly because people tend to dine out more on weekends.

```python
# Count plot: Smoker vs Non-Smoker
sns.countplot(x='smoker', data=tips)
plt.title('Count Plot: Smoker vs Non-Smoker')
plt.show()
```

Interpretation : The count plot shows that there are more non-smokers (No) than smokers (Yes) in the dataset, indicating that non-smokers are more likely to be present in the sample.

```python
# Pie chart: Proportion by day of the week
day_counts = tips['day'].value_counts()
day_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart: Proportion by Day')
plt.ylabel('')  # Hide the ylabel
plt.show()
```

Interpretation : The pie chart reveals the proportion of tips recorded on different days. We observe that Saturday and Sunday make up the majority of the dataset, suggesting more dining and tipping activity on weekends.

```python
# Histogram: Distribution of total_bill
sns.histplot(tips['total_bill'], kde=True)
plt.title('Histogram: Total Bill Distribution')
plt.show()
```

INterpretation : The histogram with KDE shows that the total_bill distribution is right-skewed, meaning most bills are on the lower end, with a few larger bills. This reflects the tendency for bills to be smaller in everyday dining and larger during special events or weekends.
